```python
import time
import os
import shutil
import numpy as np
import pandas as pd
from tqdm.notebook import tqdm
```

```python
ROOT_DIR = '..'

paths = {
    'SCRIPTS': os.path.join(ROOT_DIR, 'Scripts'),
    'MODELS': os.path.join(ROOT_DIR, 'Models'),
    'DATASET': os.path.join(ROOT_DIR, 'Datasets'),
    'GRIDS': os.path.join(ROOT_DIR, 'Datasets', 'images'),
    'YOLO_V5': os.path.join(ROOT_DIR, 'Models', 'Yolo_v5'),
    'LABELLIMG': os.path.join(ROOT_DIR, 'labelImg'),
    'MASK': os.path.join(ROOT_DIR, 'Datasets', 'labels')
}
```

```python
classes = {'grid': '0', 'start': '1', 'freespace': '2', 'bot': '3'}
```

```python
i = 1
label_dir = os.path.join(paths['DATASET'], f'labels_{i}')
if not os.path.exists(label_dir):
    os.makedirs(label_dir)
    
classes_txt = {}

with open(os.path.join(paths['DATASET'], f'mask_{i}/classes.txt'), 'r') as f:
    lines  = f.readlines()
    for j ,line in enumerate(lines):
        classes_txt[str(j)] = line[:-1].lower()
print(classes_txt)
```

```python
for mask in tqdm(os.listdir(os.path.join(paths['DATASET'], f'mask_{i}'))):
    if mask != 'classes.txt':
        replacement = ""
        path = os.path.join(paths['DATASET'], f'mask_{i}', mask)
        file = open(path, "r")
        for line in file:
            line = line.strip()
            print(line)
            changes = classes[classes_txt[line[0]]] + line[1:]
            replacement = replacement + changes + "\n"
        print("replacement")
        print(replacement)
        print("*"*30)
        file.close()
        fout = open(os.path.join(paths['DATASET'], label_dir, mask), "w+")
        fout.write(replacement)
        fout.close()

with open(os.path.join(paths['DATASET'], label_dir, 'classes.txt'), "w+") as f:
    for k in classes.keys():
        f.write(k+"\n")
```
