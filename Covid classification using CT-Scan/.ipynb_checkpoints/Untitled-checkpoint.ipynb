```python
DATASET_PATH = "Dataset"
FILE_NAME = ['CT_COVID', 'CT_NonCOVID']
```

```python
# Getting data from different folders in dataset
# with respective category and saving it in Dataframe

df = pd.DataFrame(columns = ['Pixel','Label', 'Size', 'format', 'mode'])

for file in FILE_NAME:
    cat = file # label for image 
    images_dir = os.path.join(DATASET_PATH, file) # folder path
    print(images_dir)
    for image in os.listdir(images_dir):
        image_path = os.path.join(images_dir, image) #image path
        image = Image.open(image_path)
        img_array = np.asarray(image)     # converting image to array
        df.loc[len(df.index)] = [img_array, cat, image.size, image.format, image.mode]
```

```python
from tensorflow.keras.preprocessing import image_dataset_from_directory 
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import tensorflow as tf
```

```python
datagen = ImageDataGenerator(featurewise_center=True,
                                featurewise_std_normalization=True,
                                rotation_range=20,
                                width_shift_range=0.2,
                                height_shift_range=0.2,
                                horizontal_flip=True,
                                brightness_range=[0.2,1.0],
                                zoom_range=[0.5,0.8],
                                validation_split = 0.2)

def get_data(data_path, color_mode, image_size, datagen = datagen, seed = None, subset = None):    
    return datagen.flow_from_directory(data_path,
                                        target_size= image_size,
                                        color_mode=color_mode,
                                        subset=subset,
                                        seed=seed,
                                        class_mode='binary')
```

```python
training_data = get_data('Dataset',
                        color_mode,
                        image_size,
                        datagen=datagen,
                        seed=42,
                        subset='training')

validation_data = get_data('Dataset',
                        color_mode,
                        image_size,
                        datagen=datagen,
                        seed = 42,
                        subset='validation')
```

```python
train_df = pd.DataFrame(columns=['filename', 'class'])
test_df = pd.DataFrame(columns=['filename', 'class'])
val_df = pd.DataFrame(columns=['filename', 'class'])


for file in os.listdir(os.path.join(DATASET_PATH, 'CSV')):
    if file.startswith('train'):
        df = pd.read_csv(os.path.join(DATASET_PATH, 'CSV', file))
        train_df = pd.concat([train_df, df])
    if file.startswith('test'):
        df = pd.read_csv(os.path.join(DATASET_PATH, 'CSV', file))
        test_df = pd.concat([test_df, df])
    if file.startswith('val'):
        df = pd.read_csv(os.path.join(DATASET_PATH, 'CSV', file))
        val_df = pd.concat([val_df, df])
```
